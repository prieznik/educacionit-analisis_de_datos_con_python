## Ejercicio 3
#### Del titanic.csv, ya cargado como dataframe:
### 1. Encontrar el promedio de edad de hombres y de las mujeres

```python
# primero chequeo que no haya generos que sean diferentes a male o female
dataTitanic.loc[(dataTitanic["Sex"]!="male")&(dataTitanic["Sex"]!="female")]
```

```python
promedioEdades = dataTitanic["Age"].mean()
print(f"Promedio edades\n{promedioEdades:.2f}\n")

promedioEdadesHombres = dataTitanic.loc[dataTitanic["Sex"]=="male","Age"].mean()
print(f"Promedio edades Hombres\n{promedioEdadesHombres:.2f}\n")

promedioEdadesMujeres = dataTitanic.loc[dataTitanic["Sex"]=="female","Age"].mean()
print(f"Promedio edades Mujeres\n{promedioEdadesMujeres:.2f}\n")
```

### 2. Encontrar la cantidad de sobrevivientes de nombres y las mujeres

```python
cantidadSobrevivientes = dataTitanic.loc[dataTitanic["Survived"]==0,"Survived"].count()
print(f"Cantidad de sobrevivientes totales: {cantidadSobrevivientes}")

cantidadSobrevivientesHombres = dataTitanic.loc[(dataTitanic["Survived"]==0)&(dataTitanic["Sex"]=="male"),"Survived"].count()
print(f"Cantidad de sobrevivientes totales: {cantidadSobrevivientesHombres}")

cantidadSobrevivientesMujeres = dataTitanic.loc[(dataTitanic["Survived"]==0)&(dataTitanic["Sex"]=="female"),"Survived"].count()
print(f"Cantidad de sobrevivientes totales: {cantidadSobrevivientesMujeres}")
```
